```python
import pandas as pd

# Exercise 6a)
# Load the dataset
url = 'https://raw.githubusercontent.com/peckham-daz/24-intro-to-data-science/main/data/moma_artworks_1000.csv'
df = pd.read_csv(url)

# Display the first few rows to understand the structure
print(df.head())

# Group by 'Artist' and count the number of artworks for each
artist_counts = df['Artist'].value_counts()

# Get the artist with the most artworks
top_artist = artist_counts.idxmax()
top_count = artist_counts.max()

print(f"The artist with the most artworks is {top_artist} with {top_count} artworks.")
```

```python
artist_height = df['Height (cm)']
artist_length = df['Length (cm)']
artist_depth = df['Depth (cm)']

height_length = artist_height.multiply(artist_length, fill_value=0)
volume = height_length.multiply(artist_depth, fill_value=0)


# Get the artist with the most artworks
heighest_volume_artist = volume.idxmax()
heighest_volume = volume.max()

print(f"The artist with the heighest volume is {heighest_volume}.")
```
